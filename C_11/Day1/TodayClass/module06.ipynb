# 중복 데이터

```python
df = pd.DataFrame({
    'c1':['a','a','b','a','b'],
    'c2':[1,1,1,2,2],
    'c3':[1,1,1,2,2]
})
df
```

```python
df['c1'].duplicated()   # 겹쳤는 데이터인가 0,2가 False인 이유는 그때 첨 a, b가 나온 거고 그 이후에는 중복이니까 True로 뜬다
```

```python
df[['c2','c3']].duplicated()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
# 둘 다 c2, c3열에 중복 값 없애는 거다
df[['c2','c3']].drop_duplicates()

# df.drop_duplicates(subset=['c2','c3'], inplace = True)
df
```

```python
df_car = pd.read_csv('./code/auto-mpg.csv', header= None)
df_car.columns = ['mpg', 'cylynder', 'display', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']
df_car
```

```python
mpg_to_kpl = 1.60934 / 3.78541
df_car['kpl'] = round(df_car['mpg'] * mpg_to_kpl, 2)    # 2L연비가 나온다
df_car
```

```python
df_car.dtypes
```

```python
df_car.horsepower.unique()
```

```python
df_car.info()   # horsepower에 ?가 실측치로 포함시켜버려서 실제 실측치로 뜨지 않는다
```

```python
df_car['horsepower'].replace('?',np.nan,inplace=True)   # 실제 실측치로 바꾸기 위해 ?를 nan으로 바꾼다
```

```python
df_car.info()
```

```python
df_car.dropna(subset=['horsepower'], inplace=True)
df_car.info()
```

```python
df_car['horsepower']=df_car['horsepower'].astype(float)
```

```python
df_car.horsepower.unique()
```

```python
df_car.info()
```

```python
count, bin_dividers = np.histogram(df_car['horsepower'], bins=3)    # 수학 연산을 하고(모든 것을 앞에서부터 쭉 나열해서) 크게 3가지로 나눈다
bin_dividers
```

```python
df_car['hp_bin'] = pd.cut(
    x=df_car['horsepower'],
    bins=bin_dividers,
    labels=['저출력','보통출력','고출력'],
    include_lowest=True
)
df_car.head(10)
```

# 시계열 데이터

```python
df_stock = pd.read_csv('./code/stock-data.csv')
df_stock
```

```python
df_stock.info()
```

```python
df_stock['new_date'] = pd.to_datetime(df_stock['Date']) # Date열에서 datetime형으로 받아서 new_data열을 만든다 
                                                        # -> datetime으로 만드면 컴퓨터가 알아서 년, 월, 일로 인식한다
df_stock
```

```python
df_stock.info()
```

```python
df_stock['year']=df_stock['new_date'].dt.year
df_stock['month']=df_stock['new_date'].dt.month
df_stock['day']=df_stock['new_date'].dt.day
```

```python
df_stock
```

```python
# 일별 상품 판매에 대한 매출액으로 찾을때 날짜에 따라 모든 데이터가 있어야한다
ts_ms = pd.date_range(start='2020-01-01',
                      end=None,
                      periods=6,
                      freq='MS',
                      tz='Asia/Seoul')     
# period = 6 -> 6개, freq  MS -> Month Start frequency를 dnjffh wnaus dnjffhwnrh tz = 시간을 서울 기준으로 준다
```

```python
ts_ms
```

```python
# df.to_period에 대한 설명 출처: https://wikidocs.net/159717
df_stock['date_year'] = df_stock['new_date'].dt.to_period(freq='Y') 
# -> Datetime을 Period로 만들겠다, freq = 원하는 시간 단위로 변환
```
