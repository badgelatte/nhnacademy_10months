# 중복 데이터

```python
df = pd.DataFrame({
    'c1':['a','a','b','a','b'],
    'c2':[1,1,1,2,2],
    'c3':[1,1,1,2,2]
})
df
```

```python
df['c1'].duplicated()   # 겹쳤는 데이터인가 0,2가 False인 이유는 그때 첨 a, b가 나온 거고 그 이후에는 중복이니까 True로 뜬다
```

```python
df[['c2','c3']].duplicated()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
# 둘 다 c2, c3열에 중복 값 없애는 거다
df[['c2','c3']].drop_duplicates()

# df.drop_duplicates(subset=['c2','c3'], inplace = True)
df
```

```python
df_car = pd.read_csv('./code/auto-mpg.csv', header= None)
df_car.columns = ['mpg', 'cylynder', 'display', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']
df_car
```

```python
mpg_to_kpl = 1.60934 / 3.78541
df_car['kpl'] = round(df_car['mpg'] * mpg_to_kpl, 2)    # 2L연비가 나온다
df_car
```

```python
df_car.dtypes
```

```python
df_car.horsepower.unique()
```

```python
df_car.info()   # horsepower에 ?가 실측치로 포함시켜버려서 실제 실측치로 뜨지 않는다
```

```python
df_car['horsepower'].replace('?',np.nan,inplace=True)   # 실제 실측치로 바꾸기 위해 ?를 nan으로 바꾼다
```
