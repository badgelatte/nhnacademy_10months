# 남성 여성 전체 비교 그래프

```python
df_woman =df_Covid.loc[df_Covid['category'] == '여성']
df_woman.index = df_woman.date
df_woman.drop('date',axis=1,inplace=True)
df_woman
```

```python
df_man =df_Covid.loc[df_Covid['category'] == '남성']
df_man.index = df_man.date
df_man.drop('date',axis=1,inplace=True)
df_man
```

```python
df_all = df_Covid.loc[df_Covid['category'] == '전체']
df_all.index = df_all.date
df_all.drop('date',axis=1,inplace=True)
df_all
```

```python
fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
plt.style.use('ggplot')

# axe 객체에 plot 함수로 그래프 출력
ax1.plot(df_all.NumberOfPeople, linewidth=1, color='red')
ax1.set_title('전체 감염자 수')
ax2.plot(df_man.NumberOfPeople, linewidth=1, label='남자')
ax2.plot(df_woman.NumberOfPeople, markerfacecolor='green', color='olive', linewidth=1, label='여자')
ax2.set_title('남성(red), 여성(olive) 감염자 수')

plt.show()
```

```python
df_gender = pd.concat([df_man.NumberOfPeople,df_woman.NumberOfPeople,df_all.NumberOfPeople], axis=1)
df_gender.columns = 'man','woman','all' 
df_gender.dropna(how='any',axis=0,inplace=True)
df_gender
```

```python
df_gender.info()
```

```python
plt.style.use('ggplot')
df_gender_trans = df_gender.transpose()
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)

ax.plot(df_gender_trans.loc['man'], label='남자 인원수', marker="o")
ax.plot(df_gender_trans.loc['woman'], label='여자 인원수', marker="o")
ax.plot(df_gender_trans.loc['all'], label='전체 인원수', marker="o")

ax.legend(loc='best')

plt.title('성별 감염자 수', size = 30)
plt.ylabel('감염자 수',size = 15)
plt.xlabel('날짜',size = 15)
# plt.legend(loc='best', font = 15)
plt.show()
```

```python
df_gender2 = df_gender.copy()
df_gender2.man[df_gender.man < 0] = np.NaN
df_gender2.dropna(how='any',axis=0,inplace=True)


plt.style.use('ggplot')

df_gender2[['man','woman']].plot(kind='area',stacked=True, figsize=(20,10))

plt.title('성별 감염자 수', size = 30)
plt.ylabel('감염자 수',size = 15)
plt.xlabel('날짜',size = 15)

# 범례(Legend)를 나타낸다 -> 빨간색과 파란색이 man과 woman임을 알려주는 상자를 말한다
plt.legend(loc='best', fontsize = 30)   # loc='best' => 최적의 위치를 알아서 표시해준다
plt.show()
# df_gender.woman.unique()
```

```python
df_age=pd.DataFrame()
df_age['0-9세']=df_Covid.loc[df_Covid.category == '0-9세'].groupby('date').sum()
df_age['10-19세']=df_Covid.loc[df_Covid.category == '10-19세'].groupby('date')['NumberOfPeople'].sum()
df_age['20-29세']=df_Covid.loc[df_Covid.category == '20-29세'].groupby('date')['NumberOfPeople'].sum()
df_age['30-39세']=df_Covid.loc[df_Covid.category == '30-39세'].groupby('date')['NumberOfPeople'].sum()
df_age['40-49세']=df_Covid.loc[df_Covid.category == '40-49세'].groupby('date')['NumberOfPeople'].sum()
df_age['50-59세']=df_Covid.loc[df_Covid.category == '50-59세'].groupby('date')['NumberOfPeople'].sum()
df_age['60-69세']=df_Covid.loc[df_Covid.category == '60-69세'].groupby('date')['NumberOfPeople'].sum()
df_age['70-79세']=df_Covid.loc[df_Covid.category == '70-79세'].groupby('date')['NumberOfPeople'].sum()
df_age['80세 이상']=df_Covid.loc[df_Covid.category == '80세 이상'].groupby('date')['NumberOfPeople'].sum()
df_age
```

```python
df_age2 = df_age.reset_index(inplace=False)
df_age2.date = df_age2.date.dt.to_period(freq='M')
df_age3 = df_age2.copy()
df_age3.index = df_age3.date
df_age3.drop('date',axis=1,inplace=True)

df_age3
```
