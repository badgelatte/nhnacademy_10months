```python
# h14_g4
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'NanumGothic'
import seaborn as sns
```

```python
df_welpare_raw = pd.read_spss('Koweps_hpwc14_2019_beta2.sav')
df_welpare_raw
```

```python
df_job = df_welpare_raw[['h14_eco9', 'p1402_8aq1']]
df_job.columns=['job','wage']
df_job
```

```python
df_job.isna()
```

```python
# 결측값 처리 ; https://velog.io/@cndbstlr7/%ED%8C%8C%EC%9D%B4%EC%8D%AC-%EA%B8%B0%EC%B4%88-4.-%EA%B2%B0%EC%B8%A1%EC%B9%98-%EC%B2%98%EB%A6%AC
# df_job.fillna(0.0)
df_job.dropna(how='any',inplace=True)
df_job.index = df_job.job
df_job.drop('job',axis=1,inplace=True)
df_job
```

```python
listJob = sorted(df_job.index.unique())
listJob
```

```python
listJobAverage = []
listJobMany = []
for i in listJob:
    howmany = df_job.loc[(df_job.index == i)].size
    howmany = (i,howmany)
    listJobMany.append(howmany)
    data = df_job.loc[(df_job.index == i)].mean().values
    data = (i,data[0])
    listJobAverage.append(data)
listJobMany
    
```

```python
df_JobAverage = pd.DataFrame(listJobAverage)
df_JobAverage.columns=['jobNumber','wage_average']
df_JobAverage.index=df_JobAverage['jobNumber']
df_JobAverage.drop('jobNumber',axis=1,inplace=True)
df_JobAverage
```

```python
df_JobMany = pd.DataFrame(listJobMany)
df_JobMany.columns=['jobNumber','how_many']
df_JobMany.index=df_JobMany['jobNumber']
df_JobMany.drop('jobNumber',axis=1,inplace=True)
df_JobMany
```
