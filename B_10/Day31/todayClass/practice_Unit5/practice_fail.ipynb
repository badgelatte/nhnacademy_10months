```python
df_welpare_raw.p1402_8aq1 = df_welpare_raw.p1402_8aq1.replace('NaN',np.nan)     # NaN과 NaN은 서로 비교 불가능  # https://velog.io/@inhwa1025/Python-numpy-NaN%EC%9D%84-None%EC%9C%BC%EB%A1%9C-%EB%B3%80%ED%99%98%ED%95%98%EA%B8%B0
df_welpare_raw.p1402_8aq1.fillna(0.0,inplace=True)
df_welpare_raw.p1402_8aq1
```

```python
df_gender_wage = df_welpare_raw[['p1402_8aq1','h14_g3']]
df_gender_wage.columns='wage','human'
df_gender_wage
```

```python
Woman_wage =df_gender_wage.loc[(df_gender_wage['human'] == 2.0)]
Woman_wage.columns='wage','woman'
Woman_wage
```

```python
Man_wage = df_gender_wage.loc[(df_gender_wage['human'] == 1.0)]
Man_wage.columns='wage','man'
Man_wage
```

```python
Woman_wage_average = Woman_wage.wage.mean()
Woman_wage_middle = Woman_wage.wage.median()
```

```python
Man_wage_average = Man_wage.wage.mean()
Man_wage_middle = Man_wage.wage.median()
```

```python
fig = plt.figure(figsize=(20,15))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
# plt.style.use('ggplot')
# axe 객체에 plot 함수로 그래프 출력

ax1.plot(Woman_wage, markersize=10, color='blue', linewidth=2, label='여자')
ax2.plot(Man_wage, markersize=10, color='olive', linewidth=2,label='남자')

# ax1.set_ylim(,800000)
# ax2.set_ylim(50000,800000)


plt.show()
```
