```python
# h14_g4
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
df_welpare_raw = pd.read_spss('Koweps_hpwc14_2019_beta2.sav')
df_welpare_raw
```

```python
df_birth = df_welpare_raw[['h14_g4', 'p1402_8aq1']]
df_birth.columns=['birthYear','wage']
df_birth
```

```python
df_birth.wage = df_birth.wage.replace('NaN',np.nan)     # NaN과 NaN은 서로 비교 불가능  # https://velog.io/@inhwa1025/Python-numpy-NaN%EC%9D%84-None%EC%9C%BC%EB%A1%9C-%EB%B3%80%ED%99%98%ED%95%98%EA%B8%B0
df_birth.wage.fillna(0.0,inplace=True)
df_birth
```

```python
df_birth2 = df_birth.copy()             # column = birthYear, wage
df_birth3 = df_birth.copy()             # index = wage, column = birthYear
df_birth['birthYear'] = df_birth['birthYear'].astype(int)
df_birth.index = df_birth['birthYear']
df_birth.drop('birthYear',axis=1, inplace=True)
df_birth
```

```python
df_birth3['wage'] = df_birth3['wage'].astype(int)
df_birth3.index = df_birth3['wage']
df_birth3.drop('wage',axis=1, inplace=True)
df_birth3
```

```python

df_birth2.loc[(df_welpare_raw['h14_g4'] ==1968)].wage.median()
print(df_birth.loc[(df_welpare_raw['h14_g4'] ==1968)])
df_birth2
```

```python
# df_birth.plot()
WageAverageByYear = []
WageAverageByYearFinal = []
WageAverageByYear = list(map(float,WageAverageByYear))
# birthYearList2 = df_birth.index.unique().values
birthYearList = sorted(df_welpare_raw.h14_g4.astype(int).unique())

for i in birthYearList:
    a = []
    a.append(i)
    data = df_birth.loc[(df_welpare_raw['h14_g4'] == i)].mean()
    data.replace('NaN',np.NaN)
    data.fillna(0.0,inplace=True)
    a.append(data.values)
    if (data.values) == 0.:
        a.pop()
        a.pop()
    WageAverageByYear.append(a)
WageAverageByYear

for i in WageAverageByYear:
    if len(i) > 0 :
        WageAverageByYearFinal.append(i)
WageAverageByYearFinal[0][0]
# birthYearList = birthYearList
# print(birthYearList)
```

```python
df_wageaverageYear = pd.DataFrame(WageAverageByYearFinal)
df_wageaverageYear[1] = df_wageaverageYear[1].astype(float)
df_wageaverageYear.columns = ['birthYear','average_wage']
df_wageaverageYear.index = df_wageaverageYear['birthYear']
df_wageaverageYear.drop('birthYear',axis=1,inplace=True)
df_wageaverageYear
```

```python
df_wageaverageYear.rename(columns={'average_wage':'연도별 평균 임금'},inplace=True)

df_wageaverageYear
```

```python
df_wageaverageYear.plot(kind='bar', figsize=(15,5))
# plt.xlabel('생년')
plt.show()
```
