```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_welpare_raw = pd.read_spss('Koweps_hpwc14_2019_beta2.sav')
df_welpare_raw
```

```python
df_welpare_raw.p1402_8aq1[0].dtype
```

```python
df_welpare_raw.p1402_8aq1 = df_welpare_raw.p1402_8aq1.replace('NaN',np.nan)     # NaN과 NaN은 서로 비교 불가능  # https://velog.io/@inhwa1025/Python-numpy-NaN%EC%9D%84-None%EC%9C%BC%EB%A1%9C-%EB%B3%80%ED%99%98%ED%95%98%EA%B8%B0
df_welpare_raw.p1402_8aq1.fillna(0.0,inplace=True)
df_welpare_raw.p1402_8aq1
```

```python
df_gender_wage = df_welpare_raw['p1402_8aq1']
df_gender_wage.columns='wage'
df_gender_wage
```

```python
Woman_wage =df_gender_wage.loc[(df_welpare_raw['h14_g3'] == 2.0)]
Woman_wage = pd.DataFrame(Woman_wage)

Woman_wage.columns=['wage']
Woman_wage
```

```python
Man_wage = df_gender_wage.loc[(df_welpare_raw['h14_g3'] == 1.0)]
Man_wage = pd.DataFrame(Man_wage)

Man_wage.columns=['wage']
Man_wage
```

```python
fig = plt.figure(figsize=(20,15))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
# plt.style.use('ggplot')
# axe 객체에 plot 함수로 그래프 출력

ax1.plot(Woman_wage, markersize=10, color='blue', linewidth=2, label='여자')
ax2.plot(Man_wage, markersize=10, color='olive', linewidth=2, label='남자')

# ax1.set_ylim(,800000)
# ax2.set_ylim(50000,800000)


plt.show()
```

```python
mask = Woman_wage['wage'] != 0.0
Woman_wage = Woman_wage[mask]
mask = Man_wage['wage'] != 0.0
Man_wage = Man_wage[mask]
```

```python
Man_wage_average = Man_wage.mean()
Man_wage_middle = Man_wage.median()
```

```python
Woman_wage_average = Woman_wage.mean()
Woman_wage_middle = Woman_wage.median()
```

```python
fig = plt.figure(figsize=(40,30))
ax = fig.add_subplot(2,1,1)

ax.plot(Woman_wage,linewidth=2, label='여자')
ax.plot(Man_wage, linewidth=2, label='남자')
# plt.plot(Woman_wage.index, Woman_wage.values)
# plt.show()
```

```python
# df_gender_about_wage = pd.DataFrame(Woman_wage_average,Woman_wage_middle)

plt.plot(Woman_wage_average)
plt.plot(Woman_wage_middle)
plt.plot(Man_wage_average)
plt.plot(Man_wage_middle)
```

```python
df_gender = pd.DataFrame([
        [Woman_wage_average.values, Woman_wage_middle.values],
        [Man_wage_average.values, Man_wage_middle.values]],
        columns=['average','middle'],
        index=['woman','man']
                                      )
df_gender['average'] = df_gender['average'].astype(float)
df_gender['middle'] = df_gender['middle'].astype(float)
df_gender = df_gender.transpose()
df_gender
```
