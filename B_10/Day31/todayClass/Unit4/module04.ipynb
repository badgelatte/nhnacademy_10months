```python
df_gen = pd.read_excel('./code/남북한발전전력량.xlsx')
df_gen
```

```python
df_ns = df_gen.iloc[[0,5], 3:]      # 0, 5행만 띄우고 열은 3열부터만 띄우겠다
df_ns.index = ['South','North']     # 행을 south랑 north 2개 줌
df_ns.columns = df_ns.columns.map(int)  # column을 int 형으로 바꿈!!!!!!!!!!!!!!!!!!!!!!!!!!!!
df_ns.head()
```

```python
# data = 그래프 시작점, index는 x축이고, columns는 그래프 선을 말한다 
df_ns.plot()
```

```python
# 남한과 북한의 전력량을 보여준다
df_ns_trans = df_ns.transpose()
df_ns_trans.plot()      # 선은 column, index는 x축
```

```python
df_ns_trans
```

```python
df_ns_trans.dtypes
```

```python
# 막대 그래프 모양으로 보여줌
df_ns_trans.plot(kind='bar')
```

```python
# 히스토그램 모형으로 보여줌
df_ns_trans.South = pd.to_numeric(df_ns_trans.South)    # to_numeric => 숫자형으로 바꿔줌 
df_ns_trans.North = pd.to_numeric(df_ns_trans.North)
print(df_ns_trans.dtypes)
df_ns_trans.plot(kind='hist')                           # 숫자형으로 안 바꾸고 plot할라하면 오류난다
```
