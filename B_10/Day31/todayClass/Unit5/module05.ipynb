```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
df = pd.read_excel('code/시도별 전출입 인구수.xlsx')
df.head(10)
```

```python
# fillna - 결측값
df = df.fillna(method='ffill')
df
```

```python
df = df.fillna(method='ffill')  # 결측값을 앞방향으로 채우기  # 출처: https://rfriend.tistory.com/262
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul.head(10)
```

```python
df_seoul.drop('전출지별', axis=1,inplace=True)                  # 전출지별 없애기
df_seoul.rename({'전입지별':'전입지'}, axis=1,inplace=True)     # 전입지별을 전입지로 바꿈
df_seoul.set_index('전입지',inplace=True)                       # 전입지를 index로 설정
df_seoul.head(10)
```

```python
sr_one = df_seoul.loc['경기도']
sr_one.head()
```

```python
plt.plot(sr_one)
```

```python
plt.plot(sr_one.index, sr_one.values)
plt.title('서울 -> 경기도 이동 인구')
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```

```python
plt.figure(figsize=(15, 5))         # figsize = x, y (사이즈 정하기)
plt.xticks(rotation='vertical')     # x축 말들이 겹치니까 세로로 돌려버림
plt.plot(sr_one)                    # sr_one의 그래프를 그림
plt.title('서울 -> 경기도 이동 인구')
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```

## 한번에 선들 다 그리기

```python
col_years = list(map(str, range(1970, 2018)))   # map: str을 1970~2017까지 숫자로 바꿔줘라
df_2 = df_seoul.loc[['부산광역시','경상남도','대구광역시','경상북도'], col_years]

plt.style.use('ggplot')

fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(df_2.loc['부산광역시'], label='서울->부산', marker="*")
ax.plot(df_2.loc['경상남도'], label='서울->경남', marker="*")
ax.plot(df_2.loc['대구광역시'], label='서울->대구', marker="+")
ax.plot(df_2.loc['경상북도'], label='서울->경북', marker="o")

ax.legend(loc='best')

plt.show()
```

## 면적으로 그리기

```python
col_years = list(map(str, range(1970, 2018)))   # map: str을 1970~2017까지 숫자로 바꿔줘라
df_2 = df_seoul.loc[['부산광역시','경상남도','대구광역시','경상북도'], col_years]
df_4 = df_2.transpose() # 행으로 쌓이는 데이터가 훨씬 더 처리하기 좋다

df_4.plot(kind='area', stacked=True, alpha=0.5, figsize=(20,10))

plt.title('서울-> 타시도 인구 이동', size = 30)
plt.ylabel('이동 인구수', size = 20)
plt.xlabel('기간', size = 20)
plt.legend(loc='best', fontsize = 15)
plt.show()
```
