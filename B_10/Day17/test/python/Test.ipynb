```python
# 실질적 세제곱 수 예측해서 알아보기
cube = 1.1
epsilon = 0.1 # 오차 허용 범위 1.09~1.11
guess = 0.0
increment = 0.1    # 0이 늘어날수록 더 정확히 값을 예측할 수 있다
num_guesses = 0

# epsilon 안에 찾을 때 - abs(guess**3 - cube) >= epsilon로 while문 탈출
# epsilon 안에 못 찾을 때 - guess <= cube로 while문 탈출
while abs(guess**3 - cube) >= epsilon and guess <= cube:
    guess += increment
    guess = round(guess, 2)

    num_guesses += 1    # 예측 몇번했는지 찍기
    print("guess: ", guess, ", guess의 3승: ", round(guess**3,4), "실제 오차 값: ", round(round(guess**3,4) - cube, 4))
    print()
print('num_guesses = ', num_guesses)

if abs(guess**3 - cube) >= epsilon:
    print('Failed on cobe root of', cube)
else:
    print(guess, 'us close to the cube root of', cube)
```

```python
x = 4              # 찾으려는 수
epsilon = 0.5
nums = [1,2,3,5,6,8,13,15,16,17,19,25,28,30]    # 14

low = 0             # 범위 처음 위치
prevlow = 0
high = len(nums)    # 범위 마지막 위치
prevhigh = len(nums)
number = 0
average = (low + high) / 2  # 범위 중간 값
guess = int(average)        # 실수 -> 정수

while nums[guess] - x :     # 중간 위치의 값과 x의 비교
    # 이전 low, high 백업
    # nums[guess] 와 x의 크기 비교와 상관없이 prevlow, prehigh은 low, high이 바뀌기 전에 바꿔줘야한다
    prevhigh = high
    prevlow = low

    # 범위 시작 혹은 종점 값 바꾸기
    if nums[guess] > x :                     # 범위 중간 위치 값이 x값보다 크면
        high = guess                        # 범위 마지막을 범위 중간 위치값으로 변경하여 범위를 줄인다.
    else:                                   # 범위 중간 위치 값이 x값보다 크면
        low = guess                         # 범위 마지막을 범위 중간 위치값으로 변경하여 범위를 줄인다.
    print("prevlow:", prevlow, "prevhigh:", prevhigh)
    if high == prevhigh and low == prevlow:
        break
    print("guess", guess)
    print("low", low, "high", high)
    print()

    average = (low + high) / 2
    guess = int(average)
    number += 1

print("number = ", number)
print(guess + 1, 'is close to the cobe root of', x)
```
