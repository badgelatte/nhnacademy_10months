```python
import pandas as pd
```

```python
boxoffices = pd.read_html('https://movie.datamotion.co.kr/Boxoffice')
df_boxoffice = boxoffices[1]
df_boxoffice
```

```python
df_boxoffice.columns='Rank','Movie','WWBoxOffice','DomesticBoxOffice','DomesticPercentage','Budget'
df_boxoffice
```

```python
df_boxoffice.drop(index=0,axis=1, inplace=True)
# df_boxoffice.index = df_boxoffice['rank']
# df_boxoffice.drop(columns=['rank'], inplace=True)

df_boxoffice
```

```python
year = df_boxoffice.Movie.str[-5:-1]
df_boxoffice.Movie = df_boxoffice.Movie.str[:-6]
# df_boxoffice['year'] = year
df_boxoffice.insert(2,'Year',year)
df_boxoffice
```

```python
# df_boxoffice.replace({'WWBoxOffice':{'$':'',',':''}}, inplace=True)
df_boxoffice['WWBoxOffice'] = df_boxoffice['WWBoxOffice'].str.replace('$','').str.replace(',','').astype(int)
df_boxoffice['DomesticBoxOffice'] = df_boxoffice['DomesticBoxOffice'].str.replace('$','').str.replace(',','').astype(int)
df_boxoffice['Budget'] = df_boxoffice['Budget'].str.replace('$','').str.replace(',','').astype(int)

df_boxoffice
```

```python
WWBoxOffice_Average = round(df_boxoffice.WWBoxOffice.sum()/len(df_boxoffice.WWBoxOffice), 2).astype(int)
DomesticBoxOffice_Average = round(df_boxoffice.DomesticBoxOffice.sum()/len(df_boxoffice.DomesticBoxOffice), 2).astype(int)

df_boxoffice['DomesticPercentage'] = df_boxoffice['DomesticPercentage'].str.replace('%','').astype(float)
DomesticPercentage_Average = round(df_boxoffice.DomesticPercentage.sum()/len(df_boxoffice.DomesticPercentage), 2)
Budget_Average = (df_boxoffice.Budget.sum()/len(df_boxoffice.Budget)).astype(int)
```
