```python
import pandas as pd
```

```python
df_boxoffice = pd.read_csv('code/boxoffice.csv')
df_boxoffice
```

```python
df_boxoffice.columns = df_boxoffice.iloc[0]
df_boxoffice.drop(columns=0,axis=1, inplace=True)
df_boxoffice.drop(index=0,axis=0, inplace=True)
df_boxoffice.index = df_boxoffice['순위']
df_boxoffice.drop(columns=['순위'], inplace=True)

df_boxoffice
```

```python
df_boxoffice.rename(columns={'영화':'movie','세계 흥행 수익':'world box office revenue',
                             '북미 흥행 수익':'North American box office revenue','북미 %':'North American %',
                             '제작비':'production cost'},
                    inplace=True)
df_boxoffice.index.names=['ranking']
df_boxoffice
```
