```python
import pandas as pd
```

```python
df_pop202309 = pd.read_csv('code/population202309.csv')
df_pop202309
```

```python
df_pop202309.columns = ('district','population','household','pop_household','male','female')
df_pop202309
```

```python
df_pop202309.dtypes
```

```python
df_pop202309['population'] = df_pop202309['population'].str.replace(',','').astype(int)
df_pop202309['household'] = df_pop202309['household'].str.replace(',','').astype(int)
df_pop202309['pop_household'] = df_pop202309['pop_household'].astype(int)
df_pop202309['male'] = df_pop202309['male'].str.replace(',','').astype(int)
df_pop202309['female'] = df_pop202309['female'].str.replace(',','').astype(int)
df_pop202309
```

```python
df_pop202309['male_ratio'] = round(df_pop202309.male/df_pop202309.population, 4) *100 
df_pop202309['female_ratio'] = round(df_pop202309.female/df_pop202309.population, 4) *100
df_pop202309['gender_ratio'] = round(df_pop202309.male/df_pop202309.female, 4) *100
df_pop202309 
```
