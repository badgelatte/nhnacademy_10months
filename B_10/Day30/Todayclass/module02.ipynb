```python
# df_pop202309 복사본 만들기
df_pop = df_pop202309.copy()
df_pop
```

```python
# inplace = 나한테 변경사항이 저장된다라는 뜻
# df_pop.rename(columns={'행정구역':'district','총인구수':'population'}, inplace=True)  # -> 이건 지정해서 바꾸기
df_pop.columns='district','population','household','pop_household','male','female' # -> 이건 한번에 바꾼다
```

```python
df_pop
```

```python
df_pop.info()
```

```python
# 행 이름 설정하는 방법 2가지이다
# df_pop.index=df_pop.district.to_list()
df_pop.index=df_pop['district']
df_pop
```

```python
# district라는 열을 지우는 것이기때문에 이거 2번 누르면 이미 지워졌기때문에 찾을 수 없어서 오류뜬다
# 한번만 눌러라
df_pop.drop(columns=['district'],inplace=True)
```

```python
df_pop
```

```python
df_pop.info()
```

```python
df_pop.dtypes
```

```python
# 원래 population은 String으로 수를 보기 편하게 3자리씩 ,로 끊어놔서 ,를 없애야 int형으로 바꿀 수 있다
df_pop['population']=df_pop['population'].str.replace(',','').astype(int)
df_pop
```

```python
# 남성비율, 여성비율이라는 열을 추가함
df_pop['male'] = df_pop['male'].astype(int)
df_pop['female'] = df_pop['female'].astype(int)

df_pop['male_ratio'] = round(df_pop.male / df_pop.population, 4) * 100
df_pop['female_ratio'] = round(df_pop.female / df_pop.population, 4) * 100
df_pop
```

```python
df_pop['gender_ratio'] = round(df_pop.male / df_pop.female, 4) * 100
df_pop
```

```python
# axix = 축 -> axix = 1은 열을 말한다
df_pop.drop('gender_ratio', axis=1, inplace=True)
df_pop
```

```python
df_pop.dtypes
```

```python
df_pop.population.sum()
```

```python
df_pop['household']=df_pop['household'].astype(int)
df_pop['pop_household']=df_pop['pop_household'].astype(float)
```

```python
df_pop.loc['전국'] = [
    df_pop.population.sum(),
    df_pop.household.sum(),
    round(df_pop.pop_household.mean(),2),
    df_pop.male.sum(),
    df_pop.female.sum(),
    0.0,
    0.0
]
df_pop
```

```python
# tail = 젤 밑에 5개 보여준다
df_pop.tail()
```

```python
df_pop.loc['전국','male_ratio'] = round(sum(df_pop.male)/sum(df_pop.population), 4) * 100
df_pop.loc['전국','female_ratio'] = round(sum(df_pop.female)/sum(df_pop.population), 4) * 100
df_pop
```
